```python
df['Total_Percent'] = df['Agricultural Water Use (%)'] + df['Industrial Water Use (%)'] + df['Household Water Use (%)']
df['Total_Percent'].describe
```

```python
df["Agricultural_Water_BCM"] = df["Total Water Consumption (Billion Cubic Meters)"] * df["Agricultural Water Use (%)"] / 100
df["Industrial_Water_BCM"] = df["Total Water Consumption (Billion Cubic Meters)"] * df["Industrial Water Use (%)"] / 100
df["Household_Water_BCM"] = df["Total Water Consumption (Billion Cubic Meters)"] * df["Household Water Use (%)"] / 100


```

```python
df["Sector_Sum"] = df["Agricultural_Water_BCM"] + df["Industrial_Water_BCM"] + df["Household_Water_BCM"]
df["Difference"] = df["Total Water Consumption (Billion Cubic Meters)"] - df["Sector_Sum"]
```

```python
df["Groundwater_Stress_Index"] = df["Groundwater Depletion Rate (%)"] / df["Rainfall Impact (Annual Precipitation in mm)"]
```
