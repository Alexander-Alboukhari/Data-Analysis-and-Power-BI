```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
file_path = r"C:\Users\alexa\OneDrive\ADS 3_1\Data Visualisation\Assessment 3\Task1\archive_3\simplified_coffee.csv"
coffee_data = pd.read_csv(file_path)
coffee_data
```

```python

coffee_data.info()
```

```python
num_nulls = coffee_data.isnull().sum()

num_nulls
```

```python
df=coffee_data.dropna()
```

```python
num_nulls_df = df.isnull().sum()

num_nulls_df
```

```python
df=df.drop_duplicates()
df
```

```python
print(df.shape[0])
```

```python
df_clean = df.dropna(subset=["rating", "100g_USD", "roast"])

# Create a new figure
plt.figure(figsize=(14, 7))

# Base scatterplot
sns.scatterplot(
    data=df_clean,
    x='rating',
    y='100g_USD',
    hue='roast',
    style='roast',
    palette={
        'Dark': '#1f77b4',
        'Medium-Dark': '#ff7f0e',
        'Medium': '#2ca02c',
        'Medium-Light': '#d62728',
        'Light': '#9467bd'
    },
    markers=['v', '^', 's', 'o', 'D'],
    s=35,
   
)

#  top 5 most expensive coffees
top_5_coffees = df_clean.nlargest(5, '100g_USD')

for index, row in top_5_coffees.iterrows():
    rating = row['rating']
    price = row['100g_USD']
    roast_type = row['roast']
    label = f"{roast_type} (${price:.0f})"
    plt.text(rating + 0.15, price + 2, label, fontsize=8, color='black')

# average price
avg_price = df_clean['100g_USD'].mean()
plt.axhline(y=avg_price, ls='--', color='blue', alpha=0.5)
x_position = 84.3
y_position = avg_price + 1 
label_text = f"Avg Price = ${avg_price:.2f}"
plt.text(x_position, y_position, label_text, fontsize=9, color='blue')


# Labeling and grid
plt.title('Coffee Rating vs Price per 100g by Roast Type', fontsize=16)
plt.xlabel('Rating')
plt.ylabel('Price per 100g (USD)')
plt.xticks(ticks=range(84, 99), rotation=45)
plt.grid(True, linestyle=':', linewidth=0.7, alpha=0.5)
plt.legend(title='Roast Type', loc='upper left', frameon=True)
plt.tight_layout()
plt.savefig('coffee_rating_price_annotated_final.png', dpi=300, bbox_inches='tight')
plt.show()
```
